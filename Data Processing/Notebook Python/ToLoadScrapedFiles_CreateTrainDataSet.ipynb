```python
import os
import pandas as pd

path = '/Users/Maria Vallarelli/Documents/SCRAPED/RESEARCH/'
path_det = '/Users/Maria Vallarelli/Documents/SCRAPED/DETAILS/'
path_res = '/Users/Maria Vallarelli/Documents/SCRAPED/RESEARCH/RESULT'

def load_dataset(path, path_det):
    columns = ['seller','vehicle','country','address','id_annuncio','conditional_on_price','price','link_annuncio',
                'vehicle_user_desc']
    
    df_all = pd.DataFrame(columns=columns)
    dir_listing = os.listdir(path)
    files = [fi for fi in dir_listing if fi.endswith(".csv")]
    i = 1
    for filename in files:
        file = path + filename
        df_temp = pd.read_csv(file, header=0, sep=',', encoding='utf-8', skip_blank_lines=True, usecols = columns)
        df_all = pd.concat([df_all, df_temp])

    df_all = df_all.drop_duplicates('id_annuncio')
    #print(df_all['id_annuncio'].duplicated().any())
    
    columns_det = ['Carrozzeria','id_annuncio','Alimentazione','Tipo di veicolo','Anno','Marca',
               'Modello','Porte','Posti a sedere','Tipo di cambio','Tipo di unita',
               'Colore esterno','Emissioni di CO2','Rivestimenti',
               'Colore originale','Tipo di vernice','Marce',
               'Peso a vuoto','Usato Garantito','Per neopatentati',
               'Proprietari','Cilindrata','Classe emissioni','Cilindri',
               'Veicolo per non fumatori','Consumo carburante','Tagliandi certificati']
    
    df_all_det = pd.DataFrame(columns=columns_det)
    dir_listing = os.listdir(path_det)
    files = [fi for fi in dir_listing if fi.endswith(".csv")]
    for filename in files:
        file = path_det + filename
        df_temp = pd.read_csv(file, header=0, sep=',', encoding='utf-8', skip_blank_lines=True, usecols = columns_det)
        df_all_det = pd.concat([df_all_det, df_temp])
    df_all_det = df_all_det.drop_duplicates('id_annuncio')
    #print(df_all_det['id_annuncio'].duplicated().any())
    result = pd.merge(df_all_det[['Carrozzeria','id_annuncio','Alimentazione','Tipo di veicolo','Anno','Marca',
                               'Modello','Porte','Posti a sedere','Tipo di cambio','Tipo di unita',
                               'Colore esterno','Emissioni di CO2','Rivestimenti',
                               'Colore originale','Tipo di vernice','Marce',
                               'Peso a vuoto','Usato Garantito','Per neopatentati',
                               'Proprietari','Cilindrata','Classe emissioni','Cilindri',
                               'Veicolo per non fumatori','Consumo carburante','Tagliandi certificati']],
                  df_all[['seller','vehicle','country','address','conditional_on_price','price','link_annuncio',
                            'vehicle_user_desc','id_annuncio']],how='inner', on=['id_annuncio', 'id_annuncio'])
    
    with open(path_res + "result.csv", "w", encoding="utf-8", newline='') as reference:
        result.to_csv(reference, sep=",", index=False, encoding="utf-8")
    return result
```

```python
df = pd.read_csv(path_res + "result.csv", header=0, sep=',', encoding='latin-1', skip_blank_lines=True, usecols = columns)

isNew = df['Tipo di veicolo'] == 'Nuovo'
isKM0 = df['Tipo di veicolo'] == 'KM0'
df_N = df[isNew]
df_KM0 = df[isKM0]
print(df_N.groupby(df_N['Modello'])['price'].median())

df_KM0
```

```python
df1 = pd.DataFrame(columns=['Modello', 'median'])
#  median of price calculated of new car models
df1['median'] = df_N.groupby(df_N['Modello'])['price'].median()
df1['Modello'] = df_N.groupby(df_N['Modello'])['Modello'].first()
df1.index = np.arange(1, len(df1) + 1)
```

```python
#merged KM0 cars with dataframe of median price
df_KM0_discount = pd.merge(df_KM0[['Carrozzeria','id_annuncio','Alimentazione','Tipo di veicolo','Anno','Marca',
                               'Modello','Porte','Posti a sedere','Tipo di cambio','Tipo di unita',
                               'Colore esterno','Emissioni di CO2','Rivestimenti',
                               'Colore originale','Tipo di vernice','Marce',
                               'Peso a vuoto','Usato Garantito','Per neopatentati',
                               'Proprietari','Cilindrata','Classe emissioni','Cilindri',
                               'Veicolo per non fumatori','Consumo carburante','Tagliandi certificati',
                               'seller','vehicle','country','address','conditional_on_price','price',
                               'link_annuncio','vehicle_user_desc']],
                        df1[['Modello','median']],how='inner', on=['Modello', 'Modello'])

# discount calculated compared to median of price for the same model
df_KM0_discount['discount'] = (df_KM0_discount['median'] - df_KM0_discount['price'])
df_KM0_discount
```

```python
#median discount calculated
median_discount =df_KM0_discount['discount'].median()
df_KM0_discount['median_discount'] = median_discount
#variation compared to median discount calculated
var = df_KM0_discount['discount'] - median_discount
df_KM0_discount['median_var'] = var
median_discount
```

```python
#target created by the comparison between discount and median of discount
df_KM0_discount.loc[df_KM0_discount['median_var'] < median_discount, 'target'] = '0' #under median
df_KM0_discount.loc[df_KM0_discount['median_var'] >= median_discount, 'target'] = '1' #over median
```

```python
with open(path_res + "df_KM0_discount.csv", "w", encoding="utf-8", newline='') as reference:
        df_KM0_discount.to_csv(reference, sep=",", index=False, encoding="utf-8")
```
