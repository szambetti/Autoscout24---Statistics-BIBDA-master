```python
import pandas as pd
```

```python
data = pd.read_csv("df_discount.csv") 
# Preview the first 5 lines of the loaded data 
data.head()
```

```python
mediana=data['discount'].median()
mediana
```

```python
med=data['discount']-mediana
```

```python
med
```

```python
data['variazione_su_mediana']=med
```

```python
data
```

```python
data.loc[data['variazione_su_mediana'] < 1250.0, 'equal_or_lower'] = '-1' #sotto la mediana
data.loc[data['variazione_su_mediana'] > 1250.0, 'equal_or_lower'] = '1' #sopra la mediana
data.loc[data['variazione_su_mediana'] == 1250.0, 'equal_or_lower'] = '0' #sulla mediana
```

```python
data
```

```python
data['equal_or_lower'].value_counts()
```

```python
data.to_csv('analisi_mediana.csv', index=True) 
```
